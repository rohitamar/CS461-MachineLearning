```python
import math
import numpy as np

size = 30
prob = 0.55

samples = np.random.binomial(n = 1, p = prob, size = size)
sample_mean = samples.mean()

mu = sample_mean
sigma = math.sqrt((mu * (1 - mu)) / size)

def func(p):
    pb = 0
    for i in range(501, 1001):
        pb += math.comb(1000, i) * pow(p, i) * pow(1 - p, 1000 - i)
    return pb

avg, i = 0, 0
while i < 1000:
    clt = np.random.normal(mu, sigma, 1)
    if clt[0] >= 0 and clt[0] <= 1:
        avg += func(clt[0])
        i += 1
avg = avg / 1000

print('Probability of Candidate A winning on p: ', func(0.55))
print('p_hat: ', sample_mean)
print('Probability of Candidate A winning on p_hat: ', func(sample_mean))
print('Probability of Candidate A winning on CLT sampled p_hat: ', avg)
```
